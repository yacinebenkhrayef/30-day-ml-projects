# Load the dataset

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
data
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df['label']= data.target
```

# Explore the data

```python
df
```

```python
df.describe()
```

#Preprocess the Features

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler  # or MinMaxScaler
```

```python
x=df.drop('label',axis=1)
y=df['label']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=56)
```

```python
scaler =StandardScaler()
x_train_scaled =scaler.fit_transform(x_train)
x_test_scaled =scaler.transform(x_test)
```

```python
pd.DataFrame(x_train_scaled, columns=x.columns).head()
```

```python
pd.DataFrame(x_test_scaled, columns=x.columns).head()
```

#Train a Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

###Initialize and train the Logistic Regression model

```python
model =LogisticRegression(random_state=75)
```

```python
model.fit(x_train_scaled,y_train)
```

###Make predictions on the test set

```python
y_pred= model.predict(x_test_scaled)
```

```python
y_pred
```

#Evaluate the Model

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
conf_matrix =confusion_matrix(y_test,y_pred)
conf_matrix
```

```python
precision = precision_score(y_test, y_pred)
precision
```

```python
recall = recall_score(y_test, y_pred)
recall
```

```python
f1 = f1_score(y_test, y_pred)
f1
```
